```python
import pandas as pd
```

```python
csv_file = "energy-consumption-by-source-and-region.csv"
energy_consumption_df = pd.read_csv(csv_file)
energy_consumption_df.head()
```

```python
csv_file = "FAOSTAT_data_5-2-2020.csv"
emmissions_df = pd.read_csv(csv_file)
emmissions_df.head()
```

```python
#cleaned emissions df omitted uneeded columns
new_emmissions_df = emmissions_df[['Country', 'Element', 'Item','Year','Unit','Value']].copy()
new_emmissions_df.head()
```

```python
new_energy_consumption_df = energy_consumption_df.dropna(subset=['Code'])
new_energy_consumption_df.head()
```

```python
new_energy_consumption_df['Code'].unique()
```

```python
USSR = new_energy_consumption_df[new_energy_consumption_df['Code']=='OWID_USS'].index
new_energy_consumption_df.drop(USSR , inplace=True)
new_energy_consumption_df
```

```python
World = new_energy_consumption_df[new_energy_consumption_df['Code']=='OWID_WRL'].index
new_energy_consumption_df.drop(World , inplace=True)
new_energy_consumption_df
```

```python
new_energy_consumption_df
# new_energy_consumption_df['Entity'] = new_energy_consumption_df['Entity'].replace({'Congo':'Congo, Republic Of The'})
new_energy_consumption_df['Entity'].unique()
```

```python
new_emmissions_df["Country"].unique()
```

```python
new_emmissions_df["Country"] = new_emmissions_df["Country"].replace({'China, Hong Kong SAR':'China'})
new_emmissions_df["Country"] = new_emmissions_df["Country"].replace({'China, Macao SAR':'China'})
new_emmissions_df["Country"] = new_emmissions_df["Country"].replace({'China, mainland':'China'})
new_emmissions_df["Country"] = new_emmissions_df["Country"].replace({'China, Taiwan Province of':'China'})
new_emmissions_df["Country"] = new_emmissions_df["Country"].replace({'Ethiopia PDR':'Ethiopia'})
```

```python
USSR2 = new_emmissions_df[new_emmissions_df['Country']=='USSR'].index
new_emmissions_df.drop(USSR2 , inplace=True)
```

```python
Yugoslav = new_emmissions_df[new_emmissions_df['Country']=='Yugoslav SFR'].index
new_emmissions_df.drop(Yugoslav , inplace=True)
```

```python
new_energy_consumption_df.head()
```

```python
new_energy_consumption_df.rename(columns = {"Entity":"Country"}, inplace=True)
new_energy_consumption_df = new_energy_consumption_df[['Country', 'Year', 'Oil (terrawatt-hours)','Coal (terrawatt-hours)','Gas (terrawatt-hours)','Hydropower (terrawatt-hours)','Nuclear (terrawatt-hours)','Solar (terrawatt-hours)','Wind (terrawatt-hours)','Other renewables (terawatt-hours)']].copy()
new_energy_consumption_df.head()
```
